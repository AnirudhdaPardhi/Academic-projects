# Data Preparation

Importing the required modules and reading the csv file in pandas

```python
# Importing required modules
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np 
import math
import seaborn as sns
```

```python
#importing the csv file
abc = pd.read_csv('risk_factors_cervical_cancer.csv', header=0)
```

```python
#Copying orignal dataframe to another for reference
df=abc.copy()
```

```python
#checking first 5 records of dataframe using head function
df.head()
```

# Data Cleaning and preprocessing
We are initially checking datatypes of attribues if they match with description of orignal data source.

```python
#checking the datatypes of attributes
df.dtypes
```

# Statastical description of data

Here we are checking statastics for all the attribues of dataframe

```python
df.describe()
```

# Data insight 
To get more insight about our dataset we are analysing what different types of values are present in each attribute. 

```python
#checking for unique values present in all attributes 
for a in df.columns.difference(['Age']):
    print(a)
    print(df[a].value_counts(dropna = False))
```

### Dropping insignificant attributes
We are dropping 2 columns which were insignificant and had lots of missing values

```python
df.drop(['STDs: Time since first diagnosis','STDs: Time since last diagnosis'],axis=1,inplace=True)
```

### Replacing corrupt data value with null values
We are replacing character '?' with null values for all numerical attribues.

```python
df = df.replace('?', np.NaN)
```

# Appropriate datatype conversion
We are converting the object datatype attributes to numerical float datatype which were continuous in nature and checking if all the conversion are properly applied. All the categorical datatypes are divided into integer 0 or 1 where 0 indicates No and 1 indicates Yes. 

```python
#Converting the appropriate attribues datatype from object to numeric
df_cat=df.select_dtypes(include="object")
df_cat1=df_cat.apply(pd.to_numeric)
df_cat1.columns
```

```python
#Checking if object datatype has been successfully converted into float datatype
df_cat1.dtypes
```

```python
#Checking for (categorical)attributes with integer datatypes
df_num=df.select_dtypes(include="number")
df_num.dtypes
```

```python
#Concatinating the float dataype(continuous variable) dataframe and integer dataype(categorical variable) dataframe 
df=pd.concat([df_cat1,df_num],axis=1,join="inner")
```

# Handling null values
In order to handle null values across dataset we are preparing 2 types of list:
* List of numerical continuous attributes
* List of categorical attributes

We are filling missing values in **categorical variables** with their respective feature's **mode** by looping through list of numerical continuous attributes.

We are filling missing values in **continuous variables** with their respective feature's **mean** by looping through list of numerical continuous attributes.

```python
#List of numerical continuous attributes
continuous_df = ['Age', 'Number of sexual partners', 'First sexual intercourse','Num of pregnancies', 'Smokes (years)',
                'Smokes (packs/year)','Hormonal Contraceptives (years)','IUD (years)','STDs (number)']
```

```python
#List of categorical attributes
categorical_df = ['Smokes','Hormonal Contraceptives','IUD','STDs','STDs:condylomatosis','STDs:cervical condylomatosis',
                  'STDs:vaginal condylomatosis','STDs:vulvo-perineal condylomatosis', 'STDs:syphilis',
                  'STDs:pelvic inflammatory disease', 'STDs:genital herpes','STDs:molluscum contagiosum', 'STDs:AIDS', 
                  'STDs:HIV','STDs:Hepatitis B', 'STDs:HPV', 'STDs: Number of diagnosis','Dx:Cancer', 'Dx:CIN', 
                  'Dx:HPV', 'Dx', 'Hinselmann', 'Schiller','Citology', 'Biopsy']
```

```python
#Filling missing values in categorical variables
for i in categorical_df:
    
    df[i] = df[i].fillna(1.0)
```

```python
#Filling missing values in continuous variables
for i in continuous_df:
    j = round(df[i].mean(),1)
    df[i] = df[i].fillna(j)
```

### Re-checking null values presence

```python
#Checking for presence of null values in dataframe
df.isnull().sum()
```
